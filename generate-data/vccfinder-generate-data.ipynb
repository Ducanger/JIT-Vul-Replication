```python
import pandas as pd
import pickle
import os
```

```python
PROJECT = "vccfinder"
VCC_FINDER_CSV = "raw-data/VCCFinder.csv"
TRAIN_TEST_SPLIT_MODE = "cross_project" # by_time/cross_project
TRAIN_TEST_SPLIT_DIR = "raw-data/split_data"
OUTPUT_DIR = f"../data/{PROJECT}/{TRAIN_TEST_SPLIT_MODE}"
```

```python
# VCC metadata
df = pd.read_csv(VCC_FINDER_CSV)
df = df.drop(columns=['index'])
df
```

```python
def get_id(data_path='raw-data/split_data', type='by_time'):
    with open(f"{data_path}/{type}/train_ids.txt") as file:
        train_id = [line.rstrip() for line in file]
    with open(f"{data_path}/{type}/test_ids.txt") as file:
        test_id = [line.rstrip() for line in file]
    return train_id, test_id

train_id, test_id = get_id(data_path=TRAIN_TEST_SPLIT_DIR, type=TRAIN_TEST_SPLIT_MODE) 
train, test = df.loc[df['commit_id'].isin(train_id)], df.loc[df['commit_id'].isin(test_id)]
train, test = train.reset_index().drop(columns=['index']), test.reset_index().drop(columns=['index'])

if not os.path.exists(OUTPUT_DIR): 
    os.makedirs(OUTPUT_DIR)
with open(os.path.join(OUTPUT_DIR, "features_train.pkl"), 'wb') as f:
    pickle.dump(train, f)
with open(os.path.join(OUTPUT_DIR, "features_test.pkl"), 'wb') as f:
    pickle.dump(test, f)
```
